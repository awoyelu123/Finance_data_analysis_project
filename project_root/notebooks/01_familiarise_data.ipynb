```python
import pandas as pd

df = pd.read_csv("../data/loan_payments.csv", header=None)
print("CSV shape:", df.shape)
```

```python
len(columns)
```

```python
if df.shape[1] == len(columns) + 1:
    df = df.iloc[:, 1:]

df.columns = columns
```

```python
df.head()
```

```python
import numpy as np
import pandas as pd

# df already loaded
# 1) Drop the first row if it's the dummy header-like row
df = df.iloc[1:].reset_index(drop=True)

# 2) Convert "\N" to real NaN
df = df.replace(r"\\N", np.nan, regex=True)

df.head()
```

```python
df = df.drop(columns=["index", "Unnamed: 0"], errors="ignore")
```

```python
df.to_csv("../data/loan_payments.csv", index=False)
```
